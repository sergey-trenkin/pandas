```python
for column in new_columns:
    new_list = []
    try:
        for i in range(1, 23):
            new_list += df.loc[:, f'{column}.{i}'][:950].tolist()
            df = df.drop([f'{column}.{i}'], axis = True)
    except:
        while len(new_list) != 20900:
            new_list += arr
    result = df[f'{column}'][:950].tolist() + new_list
    df[f'{column}'] = result
```
